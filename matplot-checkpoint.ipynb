```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('ggplot')
a=np.arange(1,50,0.1)
b=np.sin(a)
plt.plot(a,b,'+',color='blue',label='sin')
plt.plot(a,np.cos(a),'>',color='green',label='cos')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sample-graph')
plt.legend() # if we have labels in plot then use legend or it will give error
plt.show()
```

```python
x=[2,5,3,6,1,4]
y=[9,5,8,4,2,12]
plt.scatter(x,y,marker='x',s=150,alpha=1,color='blue',label='points')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('grah')
plt.legend()
plt.show()
```

```python
p=[4,23,4,5,7,6,8,9,12,23,14,15,12,19,17,18,19,24,25,26,27,22,21,19,10]
r=[1,10,20,30]
plt.hist(p,r,histtype='bar',rwidth=0.4)
plt.show()
```

```python
x=['india','australia','uk','usa','japan','china']
y=[9,6,5,10,8,9]
plt.bar(x,y)
plt.show()
```

```python
x=['india','australia','uk','usa','japan','china']
y=[9,6,5,10,8,9]
plt.pie(y,labels=x,shadow=True,explode=(0.1,0,0,0.20,0,0),autopct='%0.2f',colors=['b','g','y','c','w','m'])
plt.legend(loc='lower right')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[12,3,6,5,3]
y1=[9,4,7,8,4]
y2=[8,9,9,6,7]
plt.stackplot(x,y,y1,y2)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[12,3,6,5,3]
p=plt.subplot(2,2,1)
p.plot(x,y)
p1=plt.subplot(2,2,2)
p1.scatter(x,y)
p2=plt.subplot(223)
p2.bar(x,y)
p3=plt.subplot(2,2,4)
p3.pie(y)
plt.show()
```
