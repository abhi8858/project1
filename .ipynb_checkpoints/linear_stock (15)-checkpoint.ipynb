```python
start=datetime.datetime(2009,1,1)
stop=datetime.datetime.now()
dt=DataReader('msft','tiingo',start,stop,access_key='911ee28d70118f9cea5a84d2b8f1436fa32d3116')
```

```python
dt.reset_index(inplace=True)
dt.columns
```

```python
dt.set_index('date',inplace=True)
```

```python
dt.columns
```

```python
dt=dt[['adjClose', 'adjHigh', 'adjLow', 'adjOpen', 'adjVolume']]
```

```python
dt['adjClose'].plot()
```

```python
no_days=20
x=dt.drop('adjClose',axis=1)
y=dt['adjClose'].shift(-no_days).dropna()
```

```python
x.shape,y.shape
```

```python
x1=x[:-no_days]
x2=x[-no_days:]
x1.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_ts,y_tr,y_ts=train_test_split(x1,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
algo=LinearRegression()
algo.fit(x_tr,y_tr) #it will genrate m(slope ,coef) and c y-intercept
algo.score(x_ts,y_ts) # coeffient od determination
```

```python
from sklearn.metrics import r2_score
r2_score(y_ts,algo.predict(x_ts))#its also coefficent of determination
```

```python
from sklearn.preprocessing import scale
from sklearn.neighbors import KNeighborsRegressor
algo1=KNeighborsRegressor()
algo1.fit(scale(x_tr),y_tr) 
algo1.score(scale(x_ts),y_ts)
```

```python
from sklearn.svm import SVR
algo2=SVR()
algo2.fit(scale(x_tr),y_tr)
algo2.score(scale(x_ts),y_ts)
```

```python
from sklearn.tree import DecisionTreeRegressor
algo3=DecisionTreeRegressor()
algo3.fit(x_tr,y_tr) 
algo3.score(x_ts,y_ts)
```

```python
from sklearn.ensemble import RandomForestRegressor
algo4=RandomForestRegressor()
algo4.fit(x_tr,y_tr)
algo4.score(x_ts,y_ts)
```
